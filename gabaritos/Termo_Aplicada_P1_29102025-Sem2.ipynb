### 1ª Questão (4,0 pontos). 


**Enunciado**. Um trocador de calor opera em regime permanente. De um lado, escoa amônia com vazão mássica de 0,1 kg/s, entrando a 2 °C com título x = 0,3 e saindo à mesma temperatura como vapor saturado. Do outro lado, escoa propano a 0,5 bar, entrando a 30 °C e saindo a 20 °C. Determine a taxa de destruição de exergia no trocador, em kW.

Dados: $\dot{m}_A=0{,}10\ \text{kg/s}$; $T_{A,in}=2\text{ }^\circ \text{C}$; $x_{A,in}=0{,}30$; $x_{A,out}=1{,}0$; $p_P=0{,}5\ \text{bar}$, $T_{P,in}=30 ^\circ \text{C}$, $T_{P,out}=20 ^\circ\mathrm{C}$; $T_0=298{,}15\ \text{K}$.

Balanço de energia:

$$
    \dot W = \sum_{\text{ent}} \dot m h - \sum_{\text{sai}} \dot m  h +\sum \dot Q_i - \frac{dE}{dt}$$


Balanço de exergia:

$$
   \dot W = \sum_{\text{ent}} \dot m [(h-T_o s) - (h_0-T_o s_0)] - \sum_{\text{sai}} \dot m  [(h-T_o s) - (h_0-T_o s_0)] +\sum (1-\frac{T_0}{T_i})\dot Q_i - \frac{d(E-T_0S+p_0V)}{dt} - \dot X
$$


    
Hipóteses: regime permanente, adiabático ao exterior, fluxos isobáricos.
$$
\frac{d(E-T_0S+p_0V)}{dt}=0; \quad \dot{Q}=0; \quad \dot{W}=0
$$


Neste caso específico, pelo balanço da energia, chegamos à:

$$
    \dot m_{A} (h_{A,out}-h_{A,in}) = \dot m_{P} (h_{P,in}-h_{P,out})   \implies  \dot m_{P} = \dot m_{A}\frac{(h_{A,out}-h_{A,in})}{(h_{P,in}-h_{P,out})}
$$

E pelo balanço de exergia, lembrando que $\dot m_{A} (h_{A,out}-h_{A,in}) = \dot m_{P} (h_{P,in}-h_{P,out})$, chegamos à:


$$
\dot X = T_0 \left[ \dot m_{A} (s_{A,out}-s_{A,in}) + \dot m_{P} (s_{P,out}-s_{P,in}) \right]
$$



```python
from CoolProp.CoolProp import PropsSI

# Dados da célula
m_A   = 0.10          # kg/s
T0    = 298.15        # K
T_A_in  = 2.0 + 273.15
x_A_in  = 0.30
x_A_out = 1.00
p_P     = 0.5e5       # Pa  (0,5 bar)
T_P_in  = 30.0 + 273.15
T_P_out = 20.0 + 273.15

# Propriedades (kJ/kg e kJ/(kg.K))
h_A_in  = PropsSI('H','T',T_A_in ,'Q',x_A_in ,'Ammonia')/1000.0
s_A_in  = PropsSI('S','T',T_A_in ,'Q',x_A_in ,'Ammonia')/1000.0
h_A_out = PropsSI('H','T',T_A_in ,'Q',x_A_out,'Ammonia')/1000.0
s_A_out = PropsSI('S','T',T_A_in ,'Q',x_A_out,'Ammonia')/1000.0

h_P_in  = PropsSI('H','P',p_P,'T',T_P_in ,'Propane')/1000.0
s_P_in  = PropsSI('S','P',p_P,'T',T_P_in ,'Propane')/1000.0
h_P_out = PropsSI('H','P',p_P,'T',T_P_out,'Propane')/1000.0
s_P_out = PropsSI('S','P',p_P,'T',T_P_out,'Propane')/1000.0

# Vazão do propano (balanço de energia)
m_P = m_A*(h_A_out - h_A_in)/(h_P_in - h_P_out)

# Destruição de exergia (kW)
Xdot = T0*( m_A*(s_A_out - s_A_in) + m_P*(s_P_out - s_P_in) )

# Saída simples e direta
print("=== Amônia (A) ===")
print(f"h_A_in  = {h_A_in:.3f} kJ/kg   s_A_in  = {s_A_in:.6f} kJ/(kg.K)")
print(f"h_A_out = {h_A_out:.3f} kJ/kg  s_A_out = {s_A_out:.6f} kJ/(kg.K)")
print(f"Δh_A    = {h_A_out - h_A_in:.3f} kJ/kg   Δs_A = {s_A_out - s_A_in:.6f} kJ/(kg.K)")

print("\n=== Propano (P) ===")
print(f"h_P_in  = {h_P_in:.3f} kJ/kg   s_P_in  = {s_P_in:.6f} kJ/(kg.K)")
print(f"h_P_out = {h_P_out:.3f} kJ/kg  s_P_out = {s_P_out:.6f} kJ/(kg.K)")
print(f"Δh_P    = {h_P_out - h_P_in:.3f} kJ/kg   Δs_P = {s_P_out - s_P_in:.6f} kJ/(kg.K)")

print("\n=== Resultados ===")
print(f"m_P [kg/s] = {m_P:.6f}")
print(f"Xdot [kW]  = {Xdot:.6f}")
print(f"Qdot [kW]  = {m_A*(h_A_out - h_A_in):.6f}")  # taxa de transferência de calor pelo lado da amônia
```

### 2ª Questão (3,0 pontos). 



**Enunciado**. Um ciclo Rankine opera entre as pressões de 10 kPa e 8 MPa.  A temperatura na entrada da turbina é 400 °C, e o título na saída é de 95%.  Determine a eficiência isentrópica da turbina.

Dados: $T_3 = 400 °\text{C}$; $p_3 = 8000 \text{ kPa}$; $x_4 = 0,95$; $p_4 = 10 \text{ kPa}$

Definição de eficiência isentrópica para uma turbina:

$$
    \eta_{\text{iso}} = \frac{\dot W_{\text{real}}}{\dot W_{\text{iso}}} = \frac{(h_3-h_4)}{(h_3-h_{4,s})}
$$

Conhecemos duas propriedades do estado 3 e do estado 4, então podemos calcular suas entalpias. O estado 4s, pela definição da eficiência isentrópica, tem as seguintes propriedades: $p_{4,s}=p_4$ e $s_{4,s}=s_3$.

```python
from CoolProp.CoolProp import PropsSI

# Dados
T3 = 400.0 + 273.15     # K
p3 = 8.0e6              # Pa
p4 = 1.0e4              # Pa
x4 = 0.95               # título na saída real

# Estado 3 (entrada da turbina)
h3 = PropsSI('H','P',p3,'T',T3,'Water')/1000.0          # kJ/kg
s3 = PropsSI('S','P',p3,'T',T3,'Water')/1000.0          # kJ/(kg.K)

# Estado 4s (saída isentrópica): p4, s = s3
h4s = PropsSI('H','P',p4,'S',s3*1000.0,'Water')/1000.0  # kJ/kg

# Estado 4 (saída real): p4, x = 0.95
h4 = PropsSI('H','P',p4,'Q',x4,'Water')/1000.0          # kJ/kg

# Eficiência isentrópica da turbina
eta_iso = (h3 - h4)/(h3 - h4s)

# Saída
print("Estado 3: h3 = {:.3f} kJ/kg, s3 = {:.6f} kJ/(kg.K)".format(h3, s3))
print("Estado 4s: h4s = {:.3f} kJ/kg".format(h4s))
print("Estado 4 : h4  = {:.3f} kJ/kg (x = 0.95)".format(h4))
print("Eficiência isentrópica (eta_iso) =", round(eta_iso,6))
```

### 3ª Questão (3,0 pontos). 


**Enunciado**. Um sistema de ar-condicionado, com 12 000 BTU/h de capacidade de resfriamento, opera durante 4 horas por dia.  O refrigerante é o R-134a, operando entre as pressões de 320 kPa e 1400 kPa. A tarifa elétrica é de R$ 0,80/kWh.  Adote: 12 000 BTU/h = 3,517 kW;  1 kWh = 3600 kJ. Calcule a conta mensal de energia elétrica.

Dados: $p_1 = 320 kPa$; $p_3 = 1400 kPa$

Definições do ciclo ideal: $x_1=1$; $s_2=s_1$; $p_3 = p_2$; $x_3=0$


$$
   |\dot W_{\text{comp}}| = \frac{\dot Q_l}{\text{COP}} = \dot Q_l\frac{(h_1-h_2)}{(h_1-h_3)}  
$$

Logo, a conta mensal será:

$$
  C_{\text{mes}} = |\dot W_{\text{comp}}| \times (\text{horas por dia}) \times (\text{dias por mês}) \times (\text{tarifa [R\$/kWh]})
$$

```python
from CoolProp.CoolProp import PropsSI

# dados
QL = 3.517          # kW (12.000 BTU/h)
horas = 4           # h/dia
dias  = 30
tarifa = 0.80       # R$/kWh

P1 = 320e3          # Pa
P3 = 1.4e6          # Pa

# estados (R134a, ciclo ideal)
# 1: vapor saturado a P1
h1 = PropsSI('H','P',P1,'Q',1,'R134a')/1000.0
s1 = PropsSI('S','P',P1,'Q',1,'R134a')/1000.0

# 2s: compressão isentrópica a P3
h2 = PropsSI('H','P',P3,'S',s1*1000.0,'R134a')/1000.0

# 3: líquido saturado a P3
h3 = PropsSI('H','P',P3,'Q',0,'R134a')/1000.0

# COP e potência do compressor
COP  = (h1 - h3)/(h2 - h1)
Wdot = QL / COP                     # kW

# energia e custo
E_dia = Wdot * horas                # kWh/dia
E_mes = E_dia * dias                # kWh/mês
Custo = E_mes * tarifa              # R$

# saída
print(f"h1 = {h1:.3f} kJ/kg, s1 = {s1:.6f} kJ/(kg.K)")
print(f"h2 = {h2:.3f} kJ/kg")
print(f"h3 = {h3:.3f} kJ/kg")
print(f"COP = {COP:.3f}")
print(f"Wdot [kW] = {Wdot:.3f}")
print(f"E_mes [kWh] = {E_mes:.2f}")
print(f"Custo [R$/mês] = {Custo:.2f}")
```
